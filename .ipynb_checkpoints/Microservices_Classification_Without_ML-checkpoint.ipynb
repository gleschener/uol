# Data cleaning

```python
path = "/Users/gleschen/Downloads/test_raw_file.txt"
file=open(path,'r')
lines=file.readlines()
cummulateString=''
for line in lines:
#print(newLine)
    for char in line:
# Exclude the non alphabitical data.
        if(char.isalpha()):
            cummulateString+=char
        elif(char.isspace()):
            cummulateString+=''
        elif(char.isupper()):
            char=char.lower()
            cummulateString+=char
        else:
            cummulateString+=' '
            cummulateString+=' '
print(cummulateString)
#file.close()
cummulateString+=' '
```
