```python
#importing dataset
import sklearn.datasets as skd
categories = ['sci.med', 'sci.electronics', 'comp.windows.x', 'comp.sys.ibm.pc.hardware']
news_train = skd.load_files('/Users/gleschen/Downloads/20newsgroup/train', categories = categories, encoding = 'iso-8859-15')

news_test = skd.load_files('/Users/gleschen/Downloads/20newsgroup/test', categories = categories, encoding = 'iso-8859-15')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

```python
model.fit(news_train.data, news_train.target)
labels = model.predict(news_test.data)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(news_test.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=news_train.target_names, yticklabels=news_train.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label');

```
